```python
import pandas as pd
import os

# Define the directory path and file name
directory = "/Users/HiteshVijan/Downloads"
file_path = os.path.join(directory, "crimes1823_clean.csv")

# Read the CSV file into a DataFrame
crimes_df = pd.read_csv(file_path)

# Display the first few rows of the DataFrame to verify
print(crimes_df.head())
```

```python
print(crimes_df.columns)
```

```python
crimes_df.shape
crimes_df.columns = crimes_df.columns.str.lower().str.replace(' ', '_')
```

```python
crimes_df.head()
```

```python
crimes_df.duplicated().sum()
(crimes_df.isnull().sum()/len(crimes_df))*100

```

```python
crimes_df['date'] = pd.to_datetime(crimes_df['date'], format='%m/%d/%Y %I:%M:%S %p')
crimes_df['updated_on'] = pd.to_datetime(crimes_df['updated_on'], format ='%m/%d/%Y %I:%M:%S %p')
crimes_df['block'] = crimes_df['block'].str.replace(r'\d+X+\s+', '', regex=True)
```

```python
df = crimes_df[['id','case_number','block','date','primary_type','description','location_description','arrest','domestic','latitude','longitude','community_area','updated_on']]
df.head()
df['arrest'].unique()
```

```python
df_comm = pd.read_csv('CommAreas.csv')
df_comm.columns = df_comm.columns.str.lower().str.replace(' ', '_')

df_comm.head()
```

```python
df_comm = df_comm[['area_numbe', 'community']]
df_comm.rename(columns={'area_numbe': 'community_area'}, inplace=True)
df_comm
```

```python
fmerged_df = pd.merge(df, df_comm, on='community_area', how='left')
# fmerged_df.isnull().sum()/len(fmerged_df)*100
fmerged_df.head()

```

```python
fmerged_df['year'] = fmerged_df['date'].dt.year
filtered_df = fmerged_df[(fmerged_df['year'] >= 2018) & (fmerged_df['year'] <= 2022)]
```

```python
poverty_df = pd.read_csv("poverty.csv")

poverty_df.head()
```

```python
poverty_df.columns = poverty_df.columns.str.lower().str.replace(' ', '_')

poverty_df.head()
```

```python
# filtered_df.to_csv('Chicrime_.csv', index=False)
poverty_df = poverty_df.rename(columns={'community_area_name': 'community'})
poverty_df.head()
```

```python
# Merge the DataFrames based on the "community" column
pmerged_df = pd.merge(filtered_df, poverty_df, on="community", how="left")

# Verify the merged DataFrame
pmerged_df.head()
```
